```python
#Merge,Join and Conactenate in Pandas
data1={"Emp Id":["E01","E02","E03","E04","E05","E06"],"Names":["Ram","Shyam","Rahul","Vishal","Ravi","John"],"Age":[34,36,23,44,32,36]}
data2={"Emp Id":["E01","E02","E03","E04","E05","E06"],"Salary":[45000,56000,34000,30000,50000,62000]}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
print(df1)
print(df2)
```

```python
#merge
print(pd.merge(df1,df2,on="Emp Id",how="left"))
```

```python
#concatenate
data1={"Emp Id":["E01","E02","E03","E04","E05","E06"],"Names":["Ram","Shyam","Rahul","Vishal","Ravi","John"],"Age":[34,36,23,44,32,36]}
data2={"Emp Id":["E07","E08","E09","E010","E011","E012"],"Names":["chintu","Sonu","Ranvijay","Vicky","Ravish","james"],"Age":[34,36,23,44,32,36]}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)

print(pd.concat([df1,df2]))
```

```python
#Compare DataFrames in Pandas
import pandas as pd
dict={"Fruits":["mango","apples","banana","papaya"],"Price":[100,150,50,35],"Quantity":[15,10,10,3]}
df1=pd.DataFrame(dict)
print(df1)

df2=df1.copy()

df2.loc[0,"Price"]=120
df2.loc[1,"Price"]=175
df2.loc[3,"Price"]=30
df2.loc[0,"Quantity"]=12
df2.loc[1,"Quantity"]=15
df2.loc[3,"Quantity"]=5

print(df2)
```

```python
print(df1.compare(df2,align_axis=0))
```

```python
#Pivoting
import pandas as pd
dict={"Keys":["K1","K2","K3","K4"],"Names":["John","Ben","David","Peter"],"Houses":["red","blue","green","red"],"Grades":["3rd","8th","9th","8th"]}
df=pd.DataFrame(dict)
print(df)
```

```python
print(df.pivot(index="Keys",columns="Names",values=["Houses","Grades"]))
```

```python
#Melting 
dict={"Names":["John","Ben","David","Peter"],"Houses":["red","blue","green","red"],"Grades":["3rd","8th","9th","8th"]}
df=pd.DataFrame(dict)
print(df)
```

```python
print(pd.melt(df,id_vars=["Names"],value_vars=["Houses"]))
```
